```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Data (single list)
y = df1['age'].unique()

# Create a figure and axis
fig, ax = plt.subplots()

# Initialize a line
line, = ax.plot([], [], lw=2)

# Set up plot limits
ax.set_xlim(0, len(y) - 1)
ax.set_ylim(min(y), max(y))

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Animation function: updates the line
def animate(i):
    x = list(range(i+1))  # x values are indices
    line.set_data(x, y[:i+1])  # update the data
    return line,

# Create animation object
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(y), interval=500, blit=True)

# Save the animation
ani.save('line_animation.gif', writer='imagemagick')

# You can also save it as a GIF by changing the file extension to .gif
# ani.save('line_animation.gif', writer='imagemagick')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart
months = ['January', 'February', 'March', 'April', 'May', 'June']
sales_product_a = [12000, 15000, 17000, 13000, 19000, 22000]
sales_product_b = [10000, 14000, 16000, 12000, 18000, 21000]

# Set the width of the bars
bar_width = 0.35

# Create a NumPy array for the position of the X-axis ticks
index = np.arange(len(months))

# Create the bar chart for Product A
plt.bar(index, sales_product_a, bar_width, label='Product A', color='blue')

# Create the bar chart for Product B, shifted to the right by the width of the bars
plt.bar(index + bar_width, sales_product_b, bar_width, label='Product B', color='orange')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.title('Monthly Sales Comparison: Product A vs Product B')

# Add tick labels on the X-axis
plt.xticks(index + bar_width / 2, months)

# Add a legend to the chart
plt.legend()

# Show the chart
plt.show()
```
