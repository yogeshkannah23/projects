```python
import matplotlib.pyplot as plt

# Single list of values
y = df1['age'].unique()

# Create a plot using a single list
plt.plot(y)

# Add title and labels
plt.title('Graph Using Single List')
plt.xlabel('Index')
plt.ylabel('Values')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Data (single list)
y = df1['age'].unique()

# Create a figure and axis
fig, ax = plt.subplots()

# Initialize a line
line, = ax.plot([], [], lw=2)

# Set up plot limits
ax.set_xlim(0, len(y) - 1)
ax.set_ylim(min(y), max(y))

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Animation function: updates the line
def animate(i):
    x = list(range(i+1))  # x values are indices
    line.set_data(x, y[:i+1])  # update the data
    return line,

# Create animation object
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(y), interval=500, blit=True)

# Save the animation
ani.save('line_animation.gif', writer='imagemagick')

# You can also save it as a GIF by changing the file extension to .gif
# ani.save('line_animation.gif', writer='imagemagick')

plt.show()
```
